```python
pokemon_ds = pd.read_csv("/kaggle/input/pokemon/pokemon.csv")
pokemon_ds.head()
```

```python
pokemon_ds.columns
```

```python
pokemon_ds[pokemon_ds['is_legendary'] == 0]
```

```python
poke_ranking = pokemon_ds[pokemon_ds['is_legendary'] == 0][["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "type1", "type2"]]
poke_ranking.head()
```

```python
poke_ranking["total_stat"] = poke_ranking["hp"] + poke_ranking["attack"] + poke_ranking["defense"] + poke_ranking["sp_attack"] + poke_ranking["sp_defense"] + poke_ranking["speed"]

poke_ranking.head()
```

```python
poke_ranking["type2"] = poke_ranking["type2"].fillna("")
```

```python
pd.set_option('display.max_rows', None)
poke_ranking.sort_values(["total_stat"], ascending=False).groupby(["type1", "type2"]).head()
```
